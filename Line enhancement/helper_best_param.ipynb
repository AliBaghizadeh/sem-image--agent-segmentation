```python
best_df = pd.read_csv("data/diagnostics/best results.txt", sep="\t")
best_df.head()
```

```python
best_df_all = pd.read_csv("data/diagnostics/gridsearch_panels/best_results/best_results_stats.csv")
best_df_all.head()
```

```python
# Group by config, count images, and compute std/mean ratio
config_stats = best_df_all.groupby("config").agg({"image":"count", "std": "mean", "mean":"mean"}).rename(columns = {"image":"image_count"})
```

```python
# Calculate std/mean ratio for ranking
config_stats["std_mean_ratio"] = config_stats["std"]/config_stats["mean"]
config_stats["std_mean_ratio"].sort_values(ascending=False)
```

```python
config_stats.head()
```

```python
# Sort by image count (descending) then by std/mean ratio
config_stats_ranked = config_stats.sort_values(by = ["image_count", "std_mean_ratio"], ascending=[False, False ])
config_stats_ranked
```

```python
print("Config evaluation (sorted by frequency and std/mean ratio):")
print(config_stats_ranked)
print("\nTop config (most frequent):")
print(config_stats_ranked.index[0])
```

```python
# Extract parameter substring and group by it
# Extract dirtT and M parameters
best_df_all['dirtT_M_param'] = best_df_all['config'].str.extract(r'(_dirtT[\d.]+_M\d+)')

# Group by extracted parameter
m_param_stats = best_df_all.groupby("dirtT_M_param").agg({
    "image": "count",
    "std": "mean",
    "mean": "mean"
}).rename(columns={"image": "image_count"})

m_param_stats["std_mean_ratio"] = m_param_stats["std"] / m_param_stats["mean"]
m_param_stats_ranked = m_param_stats.sort_values(by=["image_count", "std_mean_ratio"], ascending=[False, False])

print("dirtT/M parameter evaluation:")
print(m_param_stats_ranked)
```
